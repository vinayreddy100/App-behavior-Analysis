### feature engineering process

```python
# check the dataypes of variables
app.dtypes
```

```python
app["first_open"] = [parser.parse(row_date) for row_date in app["first_open"]]
app['enrolled_date'] = [parser.parse(row_data)  if isinstance(row_data, str) else row_data for row_data in app['enrolled_date']]
```

```python
app.dtypes
```

```python
app['difference'] = (app.enrolled_date - app.first_open).astype('timedelta64[h]')
```

```python
app['difference']
```

```python
plt.hist(app['difference'].dropna(), color = 'Blue')
plt.title('distribution of time since enrolled')
plt.show()
```

```python
plt.hist(app['difference'].dropna(), color = 'Blue', range = [1, 100])
plt.title('distribution of time since enrolled')
plt.show()
```

```python
app.loc[app.difference > 48,'enrolled' ] = 0
app = app.drop(columns = ['difference', 'enrolled_date', 'first_open'])
```

```python
# formatting screenlist fields
```

```python
top_screens = pd.read_csv('top_screens.csv').top_screens.values
```

```python
top_screens
```

```python
app['screen_list'] = app.screen_list.astype(str) + ','
for sc in top_screens:
    app[sc] = app.screen_list.str.contains(sc).astype(int)
    app['screen_list'] = app.screen_list.str.replace(sc+",", "")
    
app['other'] = app.screen_list.str.count(",")
app = app.drop(columns = ['screen_list'])
```

# Funnels



```python
savings_screens = ["Saving1",
                    "Saving2",
                    "Saving2Amount",
                    "Saving4",
                    "Saving5",
                    "Saving6",
                    "Saving7",
                    "Saving8",
                    "Saving9",
                    "Saving10"]
app["SavingCount"] = app[savings_screens].sum(axis=1)
app = app.drop(columns=savings_screens)

cm_screens = ["Credit1",
               "Credit2",
               "Credit3",
               "Credit3Container",
               "Credit3Dashboard"]
app["CMCount"] = app[cm_screens].sum(axis=1)
app = app.drop(columns=cm_screens)

cc_screens = ["CC1",
                "CC1Category",
                "CC3"]
app["CCCount"] =app[cc_screens].sum(axis=1)
app = app.drop(columns=cc_screens)

loan_screens = ["Loan",
               "Loan2",
               "Loan3",
               "Loan4"]
app["LoansCount"] = app[loan_screens].sum(axis=1)
app = app.drop(columns=loan_screens)
```

```python
data = data.drop(columns = 'enrolled')
```

# Model Training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data, response, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
train_identifier = X_train['user']
X_train = X_train.drop(columns = 'user')
test_identifier = X_test['user']
X_test = X_test.drop(columns = 'user')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X = StandardScaler()
```

```python
X_train2 = pd.DataFrame(sc_X.fit_transform(X_train))
```

```python
X_test2 = pd.DataFrame(sc_X.transform(X_test))
```

```python
X_train2.columns = X_train.columns.values
X_test2.columns = X_test.columns.values
```

```python
X_train2.index = X_train.index.values
X_test2.index = X_test.index.values
```

```python
X_train = X_train2
X_test = X_test2
```

# model building

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0, penalty = 'l1')
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score, classification_report
```

```python
cm = confusion_matrix(Y_test, Y_pred)
```

```python
accuracy_score(Y_test, Y_pred)
```

```python
cm
```

```python
precision_score(Y_test, Y_pred)
```

```python
recall_score(Y_test, Y_pred)
```

```python
f1_score(Y_test, Y_pred)
```

```python
df_cm = pd.DataFrame(cm, index = (0, 1), columns = (0, 1))
plt.figure(figsize = (10,7))
sns.set(font_scale=1.4)
sns.heatmap(df_cm, annot=True, fmt='g')
print("Test Data Accuracy: %0.4f" % accuracy_score(Y_test, Y_pred))
```

```python
# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = Y_train, cv = 10)
print("SVM Accuracy: %0.3f (+/- %0.3f)" % (accuracies.mean(), accuracies.std() * 2))
```
